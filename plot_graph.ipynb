```python
import numpy as np
import matplotlib.pyplot as plt
import module

import matplotlib.pyplot as plt
import networkx as nx

def outputgraph(adjacency_matrix):
    rows, cols = np.asarray(adjacency_matrix != 0).nonzero() 
    edges = [(a,b,  {'weight': adjacency_matrix[a,b]}) for a, b in zip(rows.tolist(), cols.tolist())]
    #
    gr = nx.DiGraph()
    gr.add_edges_from(edges)
    return gr


def show_graph_base(gr):
    action_node = ['green', 20]
    normal_node = ['gray', 10]
    goal_node = ['blue', 50]
    colors = [ action_node[0] for _ in range(16)] + [normal_node[0] for _ in range(len(gr.nodes())-17)] +[goal_node[0] for _ in range(1)]
    sizes = [ action_node[1] for _ in range(16)] + [normal_node[1] for _ in range(len(gr.nodes())-17)] +[goal_node[1] for _ in range(1)]
    nx.draw(gr, node_size=sizes, arrows = True,  node_color = colors, arrowsize =5)#, labels=mylabels, with_labels=True)
    plt.show()





```

```python
n = 60
esn = module.EchoStateNetwork(n = n,  spectral_radius=3, alpha = 0.1, avg_number_of_edges=2, action_nodes=np.array([j for j in range(16)]))
G = outputgraph(esn.W)
```

```python

show_graph_base(G)
```

```python
from netgraph import Graph as plotGraph
import networkx as nx

def plot_graph_cool(G):
    cmap = 'RdYlGn'
    n = G.number_of_nodes()
    colors = { node:'g'*(node == n-1)+  'k'*(16<=node and node<n-1) + 'k'*(node <16)   for node in G.nodes() }
    print(colors,n )
    plotGraph(G, node_size=1,edge_width=0.5,node_edge_width = 0.1,
        node_label_fontdict = {},arrows = True, edge_cmap=cmap, node_color=colors)

plot_graph_cool(G)
```
